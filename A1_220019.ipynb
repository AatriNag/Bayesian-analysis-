 Q2. Previously, we defined a **Python function P** to compute the probability of an event using the naive definition of probability. Generalize that function to compute the probability of events when they are
 not all equally likely. Use this new function to compute the probability of events A and B from the previous exercise.

 Hint: you can pass a third argument with the probability of each event.

```python
def P(S, A, probabilities):
    """
    Compute the probability of event A given the sample space S and the probabilities of each outcome.

    Parameters:
    S (list): Sample space, a list of outcomes.
    A (list): Event, a list of outcomes that make up the event.
    probabilities (list): Probabilities corresponding to each outcome in the sample space S.

    Returns:
    float: The probability of the event A.
    """
    if not set(A).issubset(set(S)):
        return 0

    total_probability = 0
    for outcome in A:
        index = S.index(outcome)
        total_probability += probabilities[index]

    return total_probability

# Sample space and corresponding probabilities
sample_space = ["Strawberry", "Blueberry", "Cinnamon"]
probabilities = [0.5,0.25, 0.25]  # Assuming equal probabilities for this example

# Define events
event_A = ["Strawberry"]
event_B = ["Strawberry", "Blueberry"]

# Compute probabilities of events A and B
P_A = P(sample_space, event_A, probabilities)
P_B = P(sample_space, event_B, probabilities)

# Print the results
print(f"P(A) = {P_A}")
print(f"P(B) = {P_B}")
```

Q4. Use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes,
like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. Report your obervations.

```python
# Define the SkewNormal distribution with specified parameters
mu=[1,2,3,4,5]
sigma=[2,3,4,5,6]
alpha=[2,3,2,3,2]
s=[10,100,1000]
for i in range(5):
  skew_normal_dist = pz.SkewNormal(mu[i], sigma[i], alpha[i])
  # Plot the PDF and display the moments
  dist = skew_normal_dist.plot_pdf(moments=["m", "d", "s", "k"])
  for i in s:
    print(i)
    values = skew_normal_dist.rvs(i)
    print(np.mean(values))
    print(np.std(values))


```

As the size of random samples increases we get more close to the real values of the mean and standard deviation as shown in the graph.
